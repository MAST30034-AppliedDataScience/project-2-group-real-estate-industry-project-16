```python
import pandas as pd


file_path = '../data/raw/external/schoollocations2023.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')


schools_data = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    school_info = {
        'school_name': row['School_Name'],  
        'school_type': row['School_Type'],  
        'postcode': row['Postal_Postcode']  
    }
    
    # Handle the case where school_type is "Pri/Sec"
    if school_info['school_type'] == 'Pri/Sec':
        # Add the same school as both primary and secondary
        schools_data.append({'school_name': school_info['school_name'], 'school_type': 'Primary', 'postcode': school_info['postcode']})
        schools_data.append({'school_name': school_info['school_name'], 'school_type': 'Secondary', 'postcode': school_info['postcode']})
    else:
        schools_data.append(school_info)


schools_df = pd.DataFrame(schools_data)
pivot_df = schools_df.pivot_table(index='postcode', columns='school_type', aggfunc='size', fill_value=0)

pivot_df['school_total'] = pivot_df.sum(axis=1)

# Reset the index to make it more readable and match the desired format
pivot_df = pivot_df.reset_index()


print(pivot_df)
output_file = '../data/landing/schoolscount2023.csv'
pivot_df.to_csv(output_file, index=False)
```

```python
import pandas as pd
import json

with open('../data/raw/external/schoollocations2018.json') as f:
    data = json.load(f)

# Extract relevant data from the JSON
schools_data = []
for feature in data['features']:
    properties = feature['properties']
    school_info = {
        'school_name': properties['school_name'],
        'school_type': properties['school_type'],
        'postcode': properties['address_postcode']
    }
    
    # Handle the case where school_type is "Pri/Sec"
    if school_info['school_type'] == 'Pri/Sec':
        # Add the same school as both primary and secondary
        schools_data.append({'school_name': school_info['school_name'], 'school_type': 'Primary', 'postcode': school_info['postcode']})
        schools_data.append({'school_name': school_info['school_name'], 'school_type': 'Secondary', 'postcode': school_info['postcode']})
    else:
        schools_data.append(school_info)


df = pd.DataFrame(schools_data)

# Create a pivot table to get counts by postcode and school type
pivot_df = df.pivot_table(index='postcode', columns='school_type', aggfunc='size', fill_value=0)

# Add a 'Total' column to sum across all school types
pivot_df['school_total'] = pivot_df.sum(axis=1)

# Reset the index to make it more readable and match the desired format
pivot_df = pivot_df.reset_index()


print(pivot_df)
output_file = '../data/landing/schoolscount2018.csv'
pivot_df.to_csv(output_file, index=False)
```
