```python
import pandas as pd

# Load the CSV file (adjust encoding if necessary)
file_path = '../data/raw/schoollocations2023.csv'  # Replace with the actual file path
school_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Group the data by Postal_Postcode and School_Type to count the number of schools in each postcode by type
school_counts = school_data.groupby(['Postal_Postcode', 'School_Type']).size().reset_index(name='School_Count')


total_schools_new = school_data.groupby('Postal_Postcode').size().reset_index(name='Total_Schools')
merged_data = pd.merge(school_counts, total_schools_new, on='Postal_Postcode')

total_schools_new
```

```python
output_file_path = '../data/curated/new_schools.csv' 
total_schools_new.to_csv(output_file_path, index=False)
```

```python
import json
import pandas as pd

# Load the JSON file (replace with the path to your file)
file_path_json = '../data/raw/schoollocations2018.json'  # Replace with actual file path
with open(file_path_json, 'r', encoding='ISO-8859-1') as f:
    school_data_json = json.load(f)

# Extract relevant data: postcode and school type
school_list = [
    {
        'school_name': feature['properties']['school_name'],
        'postcode': feature['properties']['postal_postcode']

    }
    for feature in school_data_json['features']
]

# Convert to DataFrame for further analysis
school_df = pd.DataFrame(school_list)

# Group by postcode and count the number of schools
total_schools_old = school_df.groupby('postcode').size().reset_index(name='total_schools')

total_schools_old
```

```python
output_file_path = '../data/curated/old_schools.csv' 
total_schools_old.to_csv(output_file_path, index=False)
```
